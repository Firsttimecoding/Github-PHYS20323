```python
def encrypt(string, key):
    tmp_string = []
    for z in range(len(string)):
        if string[z] == " ":
            tmp_string.append(" ")
        if ord(string[z]) >= 110:
            letter_out = chr(ord(string[z]) - key)
            tmp_string.append(letter_out)
        if ord(string[z]) < 110:
            letter_outt = chr(ord(string[z]) - key + 26)
            tmp_string.append(letter_outt)
    new_string = str(tmp_string)
    return new_string
sentence = encrypt("dudedude", 13)

print(sentence)
```
