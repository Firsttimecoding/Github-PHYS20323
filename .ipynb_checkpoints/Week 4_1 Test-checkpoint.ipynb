```python
for value in [2, 55, 4, 5, 12, 10, 8, 9, 22]:
    if value > 10:
        print("Value is greater than 10 (" + str(value) + ")")
    elif value == 10:
        print("Value is exactly 10")
    else:
        print("Value is less than 10 (" + str(value) + ")")
```

```python
for i in range(1,30):
    if i == 2:
        print(i)
    elif i % 2 != 0:
        prime = True
        for j in range (2,i):
            if i % j == 0:
                prime = False
        if (prime):
            print (i)
```
