```python
## import numpy:

import numpy as np

## Physical constants:

G = 6.67 * 10**(-11) # Gravatational constant in m^3 * kg^(-1) * s^(-2)
M = 1.898 * 10**27 # mass of planet in kilograms (kg)
R = 69911000 # radius of planet in meters (m)
day = 9.950 # day on planet in hours
day_s = 9.825 # sidereal day on planet in hours

## other given periods:

t_1 = 50 # in hours
t_2 = 15 # in hours
t_3 = 5  # in hours

## define a function that gives the altitude of the satellite in meters given its period in hours:

def altitude(T = float(input('Input a period in hours: '))):
    
    T_sec = T * 3600 # converting input time in hours to time in seconds
    
    return((G*M*T_sec**2/(4*np.pi**2))**(1/3) - R)

print(altitude())
```

```python
## calculation to see what value of T yields an altitude close to zero:

for i in range(0, 150000000):
    altitude(T = i/50000000)
    if altitude(T = i/50000000) >= 0:
        print(altitude(T = i/50000000), i/50000000)
        break

## I realize that this calculation is very inefficient, but it is the method that I thought of at the time. 
    
    
```
